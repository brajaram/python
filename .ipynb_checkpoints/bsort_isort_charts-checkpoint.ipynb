** Data creation and storage process **

1. I used Java Random function to generate data.  I tried to create modularized program so that the code can be re-used.  I segregated the code for number generation, sorting and file system operations so that these can be used in other programs as well.  The generated data was stored in a text file for each input starting 2000 to 5000 with a increment step of 2000 and during the sorting process I read this data from text files and load into an array.

2. Similarly, I used Java nano time to calculate the duration.  I used two arrays, one to store individual run time and the other is to store the average run time.  Once the individual runs are completed for one step (say for general random numbers, it will run 10 times for each increment step) and then the average run time will be calculated from this array and written to an output file.  This text file was used to plot the graphs through python pandas and matplotlib.

** Java files description **

GenerateRandomNumbers.java: Generate random numbers, ascending random numbers and descending random numbers and stores them in a text file.

CommonUtilsImpl.java:  This is used to write data to file, load data file into an array and print the array

BubbleSort.java: Has bubble sort implementation and it calls the sort for all three scenarios

InsertionSort.java: Has insertion sort implementation and it calls the sort for all three scenarios

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
%matplotlib inline
```

** Chart comparing Bubble Sort and Insertion Sort for random numbers **

```python
bsort = pd.read_csv(r'./generalBubbleSort_AverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
isort = pd.read_csv(r'./generalInsertionSortNano_SortAverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
```

```python
bsort['avg_milli_seconds'] = bsort['avg_time']/1000000.0
isort['avg_milli_seconds'] = isort['avg_time']/1000000.0
```

```python
fig = plt.figure(figsize=(20, 8))
gs = GridSpec(2, 2)

ax1 = plt.subplot(gs[:, 0])
labels = ['bubble sort', 'insertion sort']
alphas = [1.0,0.8]

i=0
for frame in [bsort, isort]:
    ax1.plot(frame['no_of_records'], frame['avg_time'], label=labels[i], alpha=alphas[i])
    i+=1

plt.title('Bubble Sort vs Insert Sort for random numbers')
plt.legend()
plt.xlim(2000,50000)

ax2 = plt.subplot(gs[0, 1])
ax2.plot(bsort['no_of_records'], bsort['avg_milli_seconds'], marker='o')
plt.ylim(1,3500)
plt.title('Bubble Sort')

ax3 = plt.subplot(gs[1, 1])
ax3.plot(isort['no_of_records'], isort['avg_milli_seconds'], marker='o')
plt.ylim(1,280)
plt.title('Insertion Sort')
```

Both bubble sort and insertion sort are quadratic.  It is not clearly visible for insertion sort in the comparison plot, I plotted the bubble and insert sort separately on the right hand side.  It shows that the bubble and insert sorts are quadratic.  The comparison chart clearly shows that the insertion sort performs a lot better than bubble sort, approximately factor of 10x.

** Ascending Random Number Comparison Chart **

```python
bsortasc = pd.read_csv(r'./ascendingBubbleSortNano_AverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
isortasc = pd.read_csv(r'./ascendingInsertionSortNano_SortAverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
```

```python
fig = plt.figure(figsize=(20, 8))
gs = GridSpec(2, 2)

ax1 = plt.subplot(gs[:, 0])
labels = ['bubble sort', 'insertion sort']
alphas = [1.0,0.8]

i=0
for frame in [bsortasc, isortasc]:
    ax1.plot(frame['no_of_records'], frame['avg_time'], label=labels[i], alpha=alphas[i])
    i+=1

plt.title('Bubble Sort vs Insert Sort for ascending random numbers')
plt.legend()
plt.xlim(2000,50000)

ax2 = plt.subplot(gs[0, 1])
ax2.plot(bsortasc['no_of_records'], bsortasc['avg_time'], marker='o')
plt.title('Bubble Sort')

ax3 = plt.subplot(gs[1, 1])
ax3.plot(isortasc['no_of_records'], isortasc['avg_time'], marker='o')
plt.title('Insertion Sort')
labels = ['bubble sort', 'insertion sort']
```

In case of ascending random number, insertion sort performs better as it skips the while loop altogether and hence this is a best case scenario for insertion sort and it is linear.  However, bubble sort still has to go through two for loops and it is not performing well in this scenario as well.

** Descending Random Number Comparison Chart **

```python
bsortdesc = pd.read_csv(r'./descendingBubbleSortNano_AverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
isortdesc = pd.read_csv(r'./descendingnsertionSortNano_SortAverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
```

```python
fig = plt.figure(figsize=(20, 8))
gs = GridSpec(2, 2)

ax1 = plt.subplot(gs[:, 0])
labels = ['bubble sort', 'insertion sort']
alphas = [1.0,0.8]

i=0
for frame in [bsortdesc, isortdesc]:
    ax1.plot(frame['no_of_records'], frame['avg_time'], label=labels[i], alpha=alphas[i])
    i+=1

plt.title('Bubble Sort vs Insert Sort for descending random numbers')
plt.legend()
plt.xlim(2000,50000)

ax2 = plt.subplot(gs[0, 1])
ax2.plot(bsortdesc['no_of_records'], bsortdesc['avg_time'], marker='o')
plt.title('Bubble Sort')

ax3 = plt.subplot(gs[1, 1])
ax3.plot(isortdesc['no_of_records'], isortdesc['avg_time'], marker='o')
plt.title('Insertion Sort')
labels = ['bubble sort', 'insertion sort']
```

In descending random number scenario, both the sort performs equally the same with the insertion sort having a little bit of peformance gain over bubble sort.  This is a worst case scenario for both the cases and hence both are quadratic.

** Tried Bubble Sort Optimization if the array if already sorted **

```python
bsortpreasc = pd.read_csv(r'./ascendingBubbleSortNanoPre_AverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
```

```python
labels = ['bubble sort', 'insertion sort']

fig = plt.figure()

i=0
for frame in [bsortpreasc, isortdesc]:
    plt.plot(frame['no_of_records'], frame['avg_time'], label=labels[i])
    i+=1

plt.title('Bubble Sort with presorted flag vs Insert Sort')
plt.legend()
plt.xlim(2000,50000)
plt.show()
```

For the ascending random number scenario, added a flag to see if the array is already sorted.  If the array is already sorted then the process will terminate which can be considered as a best case scenario for bubble sort.  When using this optimzation, bubble sort performs far better than insertion sort

** Conclusion **

Insertion sort performs better than the bubble sort as there is a chance of skipping the loop even if a subset is already sorted and it performs even better on sorted arrays.  Bubble sort on the other hand, if we optimize a little sort it performs very well of presorted arrays.  To conclude, Insert sort performs well compared to Bubble sort.
