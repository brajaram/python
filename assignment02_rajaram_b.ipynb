```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
%matplotlib inline
```

** Chart comparing Merge Sort, Quick Sort and Insertion Sort for random numbers **

```python
msort = pd.read_csv(r'./generalMergeSort_AverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
qsort = pd.read_csv(r'./generalQuickSort_AverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
isort = pd.read_csv(r'./generalInsertionSortNano_SortAverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
```

```python
msort['avg_milli_seconds'] = msort['avg_time']/1000000.0
qsort['avg_milli_seconds'] = qsort['avg_time']/1000000.0
isort['avg_milli_seconds'] = isort['avg_time']/1000000.0
```

```python
fig = plt.figure(figsize=(20, 8))
gs = GridSpec(2, 2)

ax1 = plt.subplot(gs[:, 0])
labels = ['merge sort', 'quick sort', 'insertion sort']
alphas = [1.0,1.0, 1.0]

i=0
for frame in [msort, qsort, isort]:
    ax1.plot(frame['no_of_records'], frame['avg_time'], label=labels[i], alpha=alphas[i])
    i+=1

plt.title('Merge Sort vs Quick Sort Vs Insertion Sort for random numbers')
plt.legend()
plt.xlim(500,10000)

ax2 = plt.subplot(gs[0, 1])
ax2.plot(msort['no_of_records'], msort['avg_milli_seconds'], marker='o')
plt.title('Merge Sort')

ax3 = plt.subplot(gs[1, 1])
ax3.plot(qsort['no_of_records'], qsort['avg_milli_seconds'], marker='o')
plt.title('Quick Sort')
```

Both bubble sort and insertion sort are quadratic.  It is not clearly visible for insertion sort in the comparison plot, I plotted the bubble and insert sort separately on the right hand side.  It shows that the bubble and insert sorts are quadratic.  The comparison chart clearly shows that the insertion sort performs a lot better than bubble sort, approximately factor of 10x.

** Ascending Random Number Comparison Chart **

```python
msortasc = pd.read_csv(r'./ascendingMergeSortNano_AverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
qsortasc = pd.read_csv(r'./ascendingQuickSortNano_AverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
isortasc = pd.read_csv(r'./ascendingInsertionSortNano_SortAverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
```

```python
fig = plt.figure(figsize=(20, 8))
gs = GridSpec(2, 2)

ax1 = plt.subplot(gs[:, 0])
labels = ['merge sort', 'quick sort', 'insertion sort']
alphas = [1.0,1.0,0.8]

i=0
for frame in [msortasc, qsortasc, isortasc]:
    ax1.plot(frame['no_of_records'], frame['avg_time'], label=labels[i], alpha=alphas[i])
    i+=1

plt.title('Merge Sort vs Quick Sort vs Insertion Sort for ascending random numbers')
plt.legend()
plt.xlim(500,10000)

ax2 = plt.subplot(gs[0, 1])
ax2.plot(msortasc['no_of_records'], msortasc['avg_time'], marker='o')
plt.title('Merge Sort')

ax3 = plt.subplot(gs[1, 1])
ax3.plot(qsortasc['no_of_records'], qsortasc['avg_time'], marker='x')
plt.title('Quick Sort')
```

In case of ascending random number, insertion sort performs better as it skips the while loop altogether and hence this is a best case scenario for insertion sort and it is linear.  However, bubble sort still has to go through two for loops and it is not performing well in this scenario as well.

** Descending Random Number Comparison Chart **

```python
msortdesc = pd.read_csv(r'./descendingMergeSortNano_AverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
qsortdesc = pd.read_csv(r'./descendingQuickSortNano_AverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
isortdesc = pd.read_csv(r'./descendingnsertionSortNano_SortAverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
```

```python
fig = plt.figure(figsize=(20, 8))
gs = GridSpec(2, 2)

ax1 = plt.subplot(gs[:, 0])
labels = ['merge sort', 'quick sort', 'insertion sort']
alphas = [1.0,1.0, 1.0]

i=0
for frame in [msortdesc, qsortdesc, isortdesc]:
    ax1.plot(frame['no_of_records'], frame['avg_time'], label=labels[i], alpha=alphas[i])
    i+=1

plt.title('Merge Sort vs Quick Sort vs Insertion Sort for descending random numbers')
plt.legend()
plt.xlim(500,10000)

ax2 = plt.subplot(gs[0, 1])
ax2.plot(msortdesc['no_of_records'], msortdesc['avg_time'], marker='o')
plt.title('Merge Sort')

ax3 = plt.subplot(gs[1, 1])
ax3.plot(qsortdesc['no_of_records'], qsortdesc['avg_time'], marker='x')
plt.title('Quick Sort')
```

In descending random number scenario, both the sort performs equally the same with the insertion sort having a little bit of peformance gain over bubble sort.  This is a worst case scenario for both the cases and hence both are quadratic.

For the ascending random number scenario, added a flag to see if the array is already sorted.  If the array is already sorted then the process will terminate which can be considered as a best case scenario for bubble sort.  When using this optimzation, bubble sort performs far better than insertion sort

** Conclusion **

Insertion sort performs better than the bubble sort as there is a chance of skipping the loop even if a subset is already sorted and it performs even better on sorted arrays.  Bubble sort on the other hand, if we optimize a little sort it performs very well of presorted arrays.  To conclude, Insert sort performs well compared to Bubble sort.
