```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
%matplotlib inline
```

** Chart comparing Merge Sort, Quick Sort and Insertion Sort for random numbers **

```python
msort = pd.read_csv(r'./generalMergeSort_AverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
qsort = pd.read_csv(r'./generalQuickSort_AverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
isort = pd.read_csv(r'./generalInsertionSortNano_SortAverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
```

```python
msort['avg_milli_seconds'] = msort['avg_time']/1000000.0
qsort['avg_milli_seconds'] = qsort['avg_time']/1000000.0
isort['avg_milli_seconds'] = isort['avg_time']/1000000.0
```

```python
fig = plt.figure(figsize=(20, 8))
gs = GridSpec(3, 2)

ax1 = plt.subplot(gs[:, 0])
labels = ['merge sort', 'quick sort', 'insertion sort']
alphas = [1.0,1.0, 1.0]

i=0
for frame in [msort, qsort, isort]:
    ax1.plot(frame['no_of_records'], frame['avg_time'], label=labels[i], alpha=alphas[i])
    i+=1

plt.title('Merge Sort vs Quick Sort Vs Insertion Sort for random numbers')
plt.legend()
plt.xlim(500,10000)

ax2 = plt.subplot(gs[0, 1])
ax2.plot(msort['no_of_records'], msort['avg_milli_seconds'], marker='o')
plt.title('Merge Sort')

ax3 = plt.subplot(gs[1, 1])
ax3.plot(qsort['no_of_records'], qsort['avg_milli_seconds'], marker='o')
plt.title('Quick Sort')

ax4 = plt.subplot(gs[2, 1])
ax4.plot(isort['no_of_records'], isort['avg_milli_seconds'], marker='o')
plt.title('Insertion Sort')
```

Merge sort and quick sort are poly lograthmic in this case and insertion sort is clearly quadratic.  To better understand merge sort and quick sort, we can take a closer look at the individual plots on the right hand side.  Quick sort performs better in this case and could be due to the value of the pivot element.  However, the performance of merge and quick sort for very smaller input sizes are negligible.

** Ascending Random Number Comparison Chart **

```python
msortasc = pd.read_csv(r'./ascendingMergeSortNano_AverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
qsortasc = pd.read_csv(r'./ascendingQuickSortNano_AverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
isortasc = pd.read_csv(r'./ascendingInsertionSortNano_SortAverageRunTime.txt', delimiter=':', header=None, names = ['name', 'no_of_records', 'avg_time'])
```

```python
fig = plt.figure(figsize=(20, 8))
gs = GridSpec(3, 2)

ax1 = plt.subplot(gs[:, 0])
labels = ['merge sort', 'quick sort', 'insertion sort']
alphas = [1.0,1.0,0.8]

i=0
for frame in [msortasc, qsortasc, isortasc]:
    ax1.plot(frame['no_of_records'], frame['avg_time'], label=labels[i], alpha=alphas[i])
    i+=1

plt.title('Merge Sort vs Quick Sort vs Insertion Sort for ascending random numbers')
plt.legend()
plt.xlim(500,10000)

ax2 = plt.subplot(gs[0, 1])
ax2.plot(msortasc['no_of_records'], msortasc['avg_time'], marker='o')
plt.title('Merge Sort')

ax3 = plt.subplot(gs[1, 1])
ax3.plot(qsortasc['no_of_records'], qsortasc['avg_time'], marker='x')
plt.title('Quick Sort')

ax4 = plt.subplot(gs[2, 1])
ax4.plot(isortasc['no_of_records'], isortasc['avg_time'], marker='x')
plt.title('Insertion Sort')
```

In case of ascending random number, insertion sort and merge sort performs better than quick sort.  The reason is insertion sort skips the while loop altogether and hence this is a best case scenario for insertion sort and it is linear.  However, quick sort is quadratic in this case.  I would argue that merge sort is lograthmic in this case as we can see from the chart.
